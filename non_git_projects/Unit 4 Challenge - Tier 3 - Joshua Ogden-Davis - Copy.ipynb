Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# NOTE: WHILE THERE ARE TECHNICALLY ONLY 32 BOROUGHS IN LONDON, "CITY OF LONDON" IS
# AN ADMINISTRATIVE AREA WITHIN LONDON THAT IS EXTREMELY SIMILAR TO A BOROUGH.
# THUS I WILL TREAT IT LIKE A BOROUGH FOR THE REMAINDER OF THE EXERCISE, MAKING
# THE TOTAL NUMBER OF "BOROUGHS" TO BE 33.

# I already know that there's more than just borough data here,
# so we're going to see more than just the borough names & codes in those columns.
# But let's make sure that there are no unexpected values there anyway...
boroughs_column_basic_values_set = set(properties_melted.iloc[0:48, 0])

all_boroughs_column_values = set(properties_melted["borough_name"])

unusual_borough_values = set(all_boroughs_column_values)

for value in all_boroughs_column_values :
    if value in boroughs_column_basic_values_set :
        unusual_borough_values.remove(value)

print(unusual_borough_values)
```

```python
# It does. So if I wanted to make an ass of you and me, I'd assume that the values
# are missing on the same rows. But, having walked to the edge of such madness,
# I shall now retreat to safer and more pedantic waters.

# The problem, simply put:
# If any column has empty values for borough_name AND borough_code
# AND avg_house_price, then it's useless and can be dropped.
# First, let's make some bools...
bool_frame = properties_melted.isna()

# Second: safety!
properties_melted_cleaned = properties_melted.copy()

# Third: loooop!
for index, row in properties_melted_cleaned.iterrows() :
    if bool_frame.loc[index, "borough_code"] and bool_frame.loc[index, "avg_house_price"] \
        and "Unnamed" in properties_melted_cleaned.loc[index, "borough_name"]:
        properties_melted_cleaned.drop(index, inplace=True)

# Let's check...
print(properties_melted_cleaned.isna().sum())
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def dfg(borough, show_values=False) :
    # Returns the ratio of average house price in "borough" between 1998 and 2018.
    # If "show_values" is True, the relevant values will also be printed.
    # 
    # First, calculate the average price in 1998.
    # We could do it all in one line, but for readability's sake, let's just segment first.
    # By which I mean, let's make a df with only the rows for that borough in that year.
    borough_1998_rows = properties_melted_cleaned[(properties_melted_cleaned["borough_name"] == borough) \
                                                 & (properties_melted_cleaned["year"] == 1998)]
    
    # Now we take the average of all the prices in that period and divide by the sample size...
    borough_1998_price_sum = borough_1998_rows["avg_house_price"].sum()
    price_1998 = borough_1998_price_sum / len(borough_1998_rows)

    # Once more, with more 2018 feeling:
    borough_2018_rows = properties_melted_cleaned[(properties_melted_cleaned["borough_name"] == borough) \
                                           & (properties_melted_cleaned["year"] == 2018)]
    borough_2018_price_sum = borough_2018_rows["avg_house_price"].sum()
    price_2018 = borough_2018_price_sum / len(borough_2018_rows)

    # Then, ratio them like it's Twitter.
    ratio = price_2018 / price_1998

    # Pentultimately, tell the people what you know! Optionally, of course.
    # By which I mean, this only prints if the show_values kwarg is set to True.
    if show_values :
        print("Borough: ", borough)
        print("Avg price in 1998: ", price_1998)
        print("Avg price in 2018: ", price_2018)
        print("Ratio: ", ratio)

    # Ultimately, tell the computer what you know.
    return ratio

# Test it, and...
dfg("England")
```

```python
# WTF??? It worked on the second try?????
# OK, let's do the whole dang thing, then...

# Am I really saving any time by making a set to keep from referencing the df on every iteration of the loop?
all_boroughs = set(properties_melted_cleaned["borough_name"])

# Now we LOOOOOOOOOOOP!
# By which I mean, for each borough, make a tuple consisting of the borough name and its ratio.
# Then put them in a list.
results_list = []
for borough in all_boroughs :
    results_list.append((borough, dfg(borough)))

# Now let's order the list by ratio so we can easily find the nth most ratio'd borough
results_list.sort(key=lambda x: x[1], reverse=True)

# Aaaand we can print it in a human-friendly way.
for item in results_list :
    print(item[0], "=", round(item[1], 2))
```

```python
# Now that I'm looking at the list, I'm realizing a couple things.

# FIRST, I don't like that the regional data is in there. The question
# was about boroughs, so let's clean up our df to have only that relevant
# information. I don't want to lose the info entirely, though, because
# it might be useful for an in-depth analysis of the problem behind the problem.

# The "codes" for boroughs all have 0 in the second slot, so let's filter based on that.
properties_regional = pd.DataFrame()

for index, row in properties_melted_cleaned.iterrows() :
    if row[1][1] != str(0):
        properties_regional = properties_regional.append(row, ignore_index=True)
        properties_melted_cleaned.drop(index, inplace=True)

# We've also already created a "results_list" that contains regional data, so
# we can clean up/separate that out as well.
# First we make a set of only borough names from our newly updated df...
boroughs_only = set(properties_melted_cleaned["borough_name"])

# Then we double-comprehend to make a new list and remove regional values from the old one.
results_list_regions = [item for item in results_list if (item[0] in boroughs_only) == False]
results_list = [item for item in results_list if item[0] in boroughs_only]

# Let's just check to make sure that all worked...

print(set(properties_melted_cleaned["borough_name"]))
print(results_list)
```

```python
# SECOND, the problem behind the problem：
# The result they wanted was to determine "desirability," but the
# only metric they asked for was the ratio of avg price from 1998-2018.
# This ratio alone doesn't actually tell me anything useful.

# For example, a super cheap place that went up A LOT might still be a cheap dump.
# Or, a borough may have much higher prices in 2018, but could potentially have
# peaked in 2015 and is entering an undesirable nosedive now.

# What does "desirable" mean, anyway?
# Desirable for whomst?
# Desirable in whatst way? 

# I think I'm overthinking this, but I will play with it a little bit,
# because education. And by "play with it," I mean I will make a new dataframe
# that includes more useful columns that could assist with more granular analyses.

# This dataframe, I shall call:
housing_analysis_1998_2018 = pd.DataFrame()

# First, obvs, let's get all our boroughs & ratios in there, since that's easy-peasy with our existing list.
unzipped_results_list = list(zip(*results_list))
housing_analysis_1998_2018["Borough"] = unzipped_results_list[0]
housing_analysis_1998_2018["1998-2019 Ratio"] = unzipped_results_list[1]

# Another thing that would be interesting to know is the starting & ending average prices.
# We calculated them before, of course, but I was still too blindly trusting of the
# instructions to save the values at that point. Foolish!
# Well, I shall be a fool no longer. I'm about to throw those vauge instructions... for a LOOP!
# This will produce a list of the average 1998 prices per borough, in the same order as our
# existing df. 
average_1998_prices = list()
for borough in unzipped_results_list[0] :
    borough_1998_rows = properties_melted_cleaned[(properties_melted_cleaned["borough_name"] == borough) \
                                                  & (properties_melted_cleaned["year"] == 1998)]
    borough_1998_price_sum = borough_1998_rows["avg_house_price"].sum()
    price_1998 = borough_1998_price_sum / len(borough_1998_rows)
    average_1998_prices.append(price_1998)

# Again! AGAIN!!!
average_2018_prices = list()
for borough in unzipped_results_list[0] :
    borough_2018_rows = properties_melted_cleaned[(properties_melted_cleaned["borough_name"] == borough) \
                                                  & (properties_melted_cleaned["year"] == 2018)]
    borough_2018_price_sum = borough_2018_rows["avg_house_price"].sum()
    price_2018 = borough_2018_price_sum / len(borough_2018_rows)
    average_2018_prices.append(price_2018)

# Now we add these lists into our df as columns.
housing_analysis_1998_2018["1998 Avg Price"] = average_1998_prices
housing_analysis_1998_2018["2018 Avg Price"] = average_2018_prices

# IMHO this dataframe is much better suited to our original question.
# Let's take another peek at the data as it now stands...
print(housing_analysis_1998_2018.describe(), "\n")
print(housing_analysis_1998_2018.corr())
```

```python
# Now let's see how the top 5 ratio'd boroughs look in comparison to the
# rest of the market (specifically, we can judge them against the output
# from .describe(). For readability, I'll do that one again here.)
print(housing_analysis_1998_2018.describe(), "\n")

# Now we print the info for the top 5 boroughs for comparison.
for i in range(5):
    print(housing_analysis_1998_2018.iloc[i], "\n")
```

```python
# We can even take a look at the trends ourselves, if we want!
# Let's look at charts for the top 5 ratios...

# Since we're going to have to pull the data in a loop, let's
# set up our list...
list_of_arrays = []

# Now we loop, adding the relevant values to our list of arrays.
for i in range(5) :
    borough = results_list[i][0]
    new_array = properties_melted_cleaned.loc[properties_melted_cleaned["borough_name"] == borough, "avg_house_price"].to_numpy()
    list_of_arrays.append(new_array)

# Now we start building our plot! Starting with the x axis.
date_range_for_plot = properties_melted_cleaned.loc[properties_melted_cleaned["borough_name"] == borough, "month"].to_numpy()

# Now we put in our y values, one at a time, from the list we created above.
for i in range(5) :
    plt.plot(date_range_for_plot, list_of_arrays[i], label=results_list[i][0])

# Then we pretty it up and show!
plt.xlabel("Year")
plt.ylabel("Average House Price (Million Pounds)")
plt.title("Housing Costs Over Time for the 5 Top Ratio Boroughs")
plt.legend()
```
